```python
#Filling in the null values in description with the headline:
for i in df_copy[df_copy['Description'].isnull()].index:
  df_copy.loc[i,'Description']=df_copy.loc[i,'Headline']
```

```python
#convert the text to lower case
Column_list=['Description','Headline']
for i in Column_list:
  df_copy[i]=df_copy[i].str.lower()
df.head()
```

```python

#Removing html tags in content:
from bs4 import BeautifulSoup
def cleanhtml(raw_html):
  # Check if raw_html is a valid type before processing
  if isinstance(raw_html, (str, bytes)):
    cleantext = BeautifulSoup(raw_html, "lxml").text
    return cleantext
  else:
    # Handle float or other unexpected types
    return raw_html

df_copy['Description'] = df_copy['Description'].apply(cleanhtml)
df_copy['Headline'] = df_copy['Headline'].apply(cleanhtml)
df.head()
```

Removing stop words:

```python
# Import stopwords with nltk.
from nltk.corpus import stopwords
stop = stopwords.words('english')

# Exclude stopwords with Python's list comprehension and pandas.DataFrame.apply.
df_copy['desc_no_stop_words'] = df_copy['Description'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))

df_copy['desc_no_stop_words']
```
