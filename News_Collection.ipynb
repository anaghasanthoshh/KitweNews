### Function to handle paginated feed entries(default 10 pages):

```python


def get_feed_entries(feed_url, pages=10):
  #This will store all feed entries collected from each page
    all_entries = []
    for page in range(1, pages + 1):
      #constructing a url including page param
        paged_url = f"{feed_url}?paged={page}"#appending page no to go through the pages in the url
        parsed_feed = feedparser.parse(paged_url)
        entries = parsed_feed.entries
        print(f'Collecting in page {page} for {feed_url}')
        if not entries:
            break  # Exit if there are no more entries
        all_entries.extend(entries)
    return all_entries
```

Id , Source , Category, Headline and Content are the values that we are currently collecting for aggregation.

```python
# CSV file path
csv_file_path = '/content/drive/My Drive/Kitwe/data_collected.csv'

print('Loading...')

# Open the CSV file for writing
with open(csv_file_path, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)

    # Write the header row
    writer.writerow(['Id','Source', 'Category', 'Headline', 'Link', 'Description', 'Date'])

    # Loop through each RSS url
    for source_name, feed_url in RSS_FEEDS.items():
        # Get paginated entries for the feed
        entries = get_feed_entries(feed_url, pages=50)
        print(f'RSS Feed done: {source_name}')

        # Process each entry and write to CSV
        if not entries:
            print(f"No entries found for {source_name}")
            continue
        for entry in entries:
            link = entry.link
            Id=entry.id if 'id' in entry else 'N/A'
            date = entry.published if 'published' in entry else 'N/A'
            description = entry.summary if 'summary' in entry else 'N/A'
            headline = entry.title if 'title' in entry else 'N/A'

            # Collect all categories/tags
            category = ', '.join(tag.term for tag in entry.tags) if 'tags' in entry and entry.tags else 'N/A'

            # Write the entry data to the CSV file
            writer.writerow([Id,source_name, category, headline, link, description, date])

print(f"Data has been saved to {csv_file_path}")
```
